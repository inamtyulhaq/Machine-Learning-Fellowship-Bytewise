```python
# Importing pandas and matplotlib 
import pandas as pd 

netflix_df = pd.read_csv("netflix_data.csv") 

netflix_subset = netflix_df[netflix_df["type"] == "Movie"] 
netflix_movies = netflix_subset[["title", "country", "genre", "release_year", "duration"]] 
short_movies = netflix_movies[netflix_movies.duration < 60]
##############

import matplotlib.pyplot as mtp 
colors = [] 
for genre in netflix_movies['genre']:
    if 'Children' in genre:
        colors.append('yellow')
    elif 'Documentaries' in genre:
        colors.append('red')
    elif 'Stand-Up' in genre:
        colors.append('blue')
    else:
        colors.append('green')
     
fig = mtp.figure(figsize=(11,16)) 
mtp.scatter(netflix_movies.release_year, netflix_movies.duration, c=colors) 
mtp.title("Movie Duration by Year of Release") 
mtp.xlabel("Release year") 
mtp.ylabel("Duration (min)") 
mtp.show() 
answer = "yes" # as in latest years, large number of movies are having shorter duration
# this number is increasing
```
